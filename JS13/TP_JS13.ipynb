# Tugas Praktikum

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
model = Sequential([
    Flatten(input_shape=(28,28)),        # Ubah gambar 28x28 menjadi vektor 784
    Dense(128, activation='relu'),       # Hidden layer 1
    Dense(64, activation='relu'),        # Hidden layer 2
    Dense(10, activation='softmax')      # Output layer (10 kelas)
])

# 3. Kompilasi model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

# 4. Latih model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)

# 5. Evaluasi model
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi pada data uji: {acc:.4f}")
```

Coba dengan beberapa parameter lain:

- Ubah jumlah neuron di hidden layer (misal: 256 dan 128).
- Tambahkan satu hidden layer lagi.
- Bandingkan akurasi dan waktu pelatihan.
- Eksperimen dengan fungsi aktivasi Sigmoid vs ReLU.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train / 255.0
X_test = X_test / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Fungsi untuk membuat dan melatih model
def train_mnist(hidden_layers=[128,64], activation='relu', epochs=10, batch_size=32):
    model = Sequential()
    model.add(Flatten(input_shape=(28,28)))
    
    # Tambahkan hidden layer sesuai daftar
    for neurons in hidden_layers:
        model.add(Dense(neurons, activation=activation))
    
    # Output layer
    model.add(Dense(10, activation='softmax'))
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    start_time = time.time()
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1, verbose=0)
    end_time = time.time()
    
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    training_time = end_time - start_time
    
    return acc, training_time

# Eksperimen beberapa konfigurasi
configs = [
    {'hidden_layers':[128,64], 'activation':'relu'},
    {'hidden_layers':[256,128], 'activation':'relu'},
    {'hidden_layers':[128,64,32], 'activation':'relu'},  # 3 hidden layer
    {'hidden_layers':[128,64], 'activation':'sigmoid'},
    {'hidden_layers':[256,128], 'activation':'sigmoid'},
]

print("Hasil eksperimen JST MNIST:")
for cfg in configs:
    acc, t = train_mnist(hidden_layers=cfg['hidden_layers'], activation=cfg['activation'])
    print(f"Hidden layers {cfg['hidden_layers']}, Activation: {cfg['activation']}, "
          f"Akurasi: {acc:.4f}, Waktu pelatihan: {t:.2f}s")
```
