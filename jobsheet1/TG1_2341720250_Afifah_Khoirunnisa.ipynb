# KEGUNAAN MASING-MASING LIBRARY
## 1. PyPREP
📌 **Fungsi:** Membersihkan data EEG (rekaman aktivitas otak)  
- Mendeteksi channel EEG yang rusak/berisik  
- Mengurangi noise secara otomatis  

👉 Cocok untuk analisis sinyal otak atau membuat sistem **Brain-Computer Interface (BCI)**  

---

## 2. SciPy
📌 **Fungsi:** Library dasar untuk perhitungan ilmiah  
- Hitung integral, statistik, dan optimisasi  
- Olah sinyal (filtering, FFT)  
- Pemrosesan citra sederhana  

👉 Dipakai luas di **data science, machine learning, dan analisis sinyal**  

---

## 3. Weights & Biases (wandb)
📌 **Fungsi:** Memantau eksperimen machine learning  
- Logging progress training (loss, akurasi, dll)  
- Visualisasi grafik real-time  
- Simpan model dan parameter  
- Bisa dipakai untuk kolaborasi tim  

👉 Berguna untuk **training ML/DL** supaya eksperimen lebih rapi dan terkontrol  

---

## 4. pyECG
📌 **Fungsi:** Analisis sinyal jantung (ECG)  
- Deteksi puncak detak jantung (**R-peak**)  
- Hitung variasi detak (**HRV**)  

👉 Cocok untuk proyek **kesehatan**, misalnya deteksi aritmia atau analisis kondisi jantung

## 1. Contoh Tindakan Melanggar Etika dan Hukum dalam Penggunaan Kecerdasan Buatan

**a. Deepfake untuk penyebaran hoaks**  
AI dapat membuat video palsu menyerupai tokoh publik yang sedang mengatakan sesuatu padahal tidak pernah dilakukan. Hal ini bisa menyesatkan masyarakat dan merusak reputasi seseorang.  
📖 Referensi: Chesney, R., & Citron, D. (2019). *Deep Fakes: A Looming Challenge for Privacy, Democracy, and National Security*. California Law Review.  

**b. Diskriminasi dalam rekrutmen kerja**  
Algoritma AI yang digunakan untuk menyaring lamaran kerja dapat bias terhadap gender, ras, atau latar belakang tertentu karena data latihnya tidak seimbang.  
📖 Referensi: Barocas, S., Hardt, M., & Narayanan, A. (2019). *Fairness and Machine Learning*.  

**c. Pelanggaran privasi data**  
Misalnya, penggunaan AI untuk mengumpulkan dan menganalisis data pribadi tanpa persetujuan pemiliknya. Hal ini melanggar aturan perlindungan data seperti GDPR (General Data Protection Regulation) di Eropa.  
📖 Referensi: Voigt, P., & Von dem Bussche, A. (2017). *The EU General Data Protection Regulation (GDPR)*.  

---

## 2. Dampak Energi dan Lingkungan terhadap Pemanfaatan Kecerdasan Buatan

**a. Dampak Energi**  
Training model AI besar (misalnya GPT atau BERT) membutuhkan daya komputasi sangat tinggi, yang berarti konsumsi energi listrik juga sangat besar. Menurut penelitian, melatih satu model NLP besar dapat menghasilkan jejak karbon setara dengan **lima kali perjalanan pulang-pergi New York–San Francisco dengan pesawat**.  
📖 Referensi: Strubell, E., Ganesh, A., & McCallum, A. (2019). *Energy and Policy Considerations for Deep Learning in NLP*.  

**b. Dampak Lingkungan**  
Konsumsi energi yang tinggi berdampak pada peningkatan emisi karbon, yang memperburuk pemanasan global. Pusat data yang digunakan untuk melatih AI juga membutuhkan pendingin khusus, yang menambah jejak lingkungan.  

##### Cara Mengatasinya (Pendapat Pribadi)
- **Optimisasi algoritma** → gunakan model yang lebih efisien, misalnya *model compression* atau *knowledge distillation*.  
- **Pemanfaatan energi terbarukan** → pusat data AI sebaiknya ditenagai oleh sumber energi ramah lingkungan (solar, angin, hidro).  
- **Efisiensi pemakaian** → tidak semua masalah butuh model raksasa; gunakan model kecil untuk kebutuhan sederhana.  
- **Kolaborasi riset hijau** → mendorong penelitian di bidang *Green AI* yang berfokus pada efisiensi energi.  

📖 Referensi tambahan: Schwartz, R., Dodge, J., Smith, N. A., & Etzioni, O. (2020). *Green AI*. Communications of the ACM.
